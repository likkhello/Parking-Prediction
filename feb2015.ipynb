```python
import json
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = 200

# read the entire file into a python array
with open('hourly.json', 'rb') as f:
    data = f.readlines()

data = json.loads(data[0])

# holidays_lst comes from playground's finding
holidays_lst = ['2014-01-01', '2015-01-01', '2016-01-01',
 '2014-05-26', '2015-05-25', '2016-05-30',
 '2014-07-04', '2015-07-03', '2015-07-04', '2016-07-04',
 '2014-09-01', '2015-09-07', '2015-09-05',
 '2014-11-27', '2014-11-28', '2015-11-26', '2015-11-27', '2016-11-24', '2016-11-25',
 '2014-12-25', '2015-12-25', '2015-12-25', '2015-12-26']
```

```python
feb16_lst = []
base = datetime(2016, 2, 1)
month = 2
while base.month == month:
    feb16_lst += [(base.strftime("%Y-%m-%d"),[0 for i in range(24)])]
    base += timedelta(days=1)
data = data.items() + feb16_lst
```

```python
for key in data:
    print key
```

```python
# len(data) = 1137
# len(raw_data) = 27288
# 27288 / 1137 = 24
raw_data = []
for key, value in data:
    for i in range(24):
        raw_data.append((key, i , value[i]))

# DataFrame contains Date, Hour, #_of_parking
data_df = pd.DataFrame(raw_data, columns=['Date', 'Hour', '#_of_parking'])

# Decompose Date to Year, Week# & Weekday
pk_df = pd.DataFrame(map(lambda x: x.isocalendar(), pd.to_datetime(data_df['Date'])), columns=['Year', 'Week#', 'Weekday'])
pk_df['Date'] = data_df['Date']
pk_df['Hour'] = data_df['Hour']
pk_df['#_of_parking'] = data_df['#_of_parking']
# pk_df.head()
```

```python
# Based on playground's finding, not include 2012 & 2013's data
data_df =pk_df[(pk_df['Date'] > '2013-12-31')]
data_df.head()
```

```python
lst_feature = []
# Extract Week# feature
for i in range(53):
    lst_feature += ['w#_'+str(i)]
    data_df['w#_'+str(i)] = map(lambda x: 1 if i == x else 0, data_df['Week#'])
# Extract Weekday feature
for i in range(1, 8):
    lst_feature += ['d#_'+str(i)]
    data_df['d#_'+str(i)] = map(lambda x: 1 if i == x else 0, data_df['Weekday'])
actual_work_day = [1, 2, 3, 4]
weekend = [6, 7]
data_df['awd'] = map(lambda x: 1 if x in actual_work_day else 0, data_df['Weekday'])
data_df['weekend'] = map(lambda x: 1 if x in weekend else 0, data_df['Weekday'])
# Extract Date feature
data_df['?holiday'] = map(lambda x: 1 if x in holidays_lst else 0, data_df['Date'])
lst_feature += ['awd','weekend','?holiday']
# Extract Hour feature
for i in range(24):
    lst_feature += ['h#_'+str(i)]
    data_df['h#_'+str(i)] = map(lambda x: 1 if i == x else 0, data_df['Hour'])
# my opinion of hourly bin
rh1 = range(0, 5) + range(21, 24)
data_df['h_0_5_21_24'] = map(lambda x: 1 if x in rh1 else 0, data_df['Hour'])
rh2 = range(5, 7)
data_df['h_5_7'] = map(lambda x: 1 if x in rh2 else 0, data_df['Hour'])
rh3 = range(7, 19)
data_df['h_7_19'] = map(lambda x: 1 if x in rh3 else 0, data_df['Hour'])
rh4 = range(19, 21)
data_df['h_19_21'] = map(lambda x: 1 if x in rh4 else 0, data_df['Hour'])
# 6 hours for each bin
data_df['h_0_6'] = map(lambda x: 1 if x in range(0, 6) else 0, data_df['Hour'])
data_df['h_6_12'] = map(lambda x: 1 if x in range(6, 12) else 0, data_df['Hour'])
data_df['h_12_18'] = map(lambda x: 1 if x in range(12, 18) else 0, data_df['Hour'])
data_df['h_18_24'] = map(lambda x: 1 if x in range(18, 24) else 0, data_df['Hour'])
lst_feature += ['h_0_5_21_24','h_5_7','h_7_19','h_19_21','h_0_6','h_6_12','h_12_18','h_18_24']
```
